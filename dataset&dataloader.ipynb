```python
class WinedataSet(Dataset):
    
    def __init__(self):
        # load data for each instance (dataset)
        
        xydata = pd.read_csv(r"C:\Users\prith\Downloads\wine.csv",dtype=np.float32, skiprows=1)
        xy = np.array(xydata)
        self.x = xy[:,1:]
        self.y = xy[:,[0]]                 #shape is (177,1) means 177 rows and 1 column that is the label.
        self.n_samples = xy.shape[0]       # xy.shape = (177,14) ; xy.shape[0] = 177 ; xy.shape[1] = 14.
        
    def __getitem__(self,index):
        return self.x[index],self.y[index]
    
    def __len__(self):
        return self.n_samples               #attributes defined in init function can be used by any other function using self.attribute_name
        
        
        
    
```

when u call the WineddataSet(), it calls the __init__ function.


when calling dataset[0], this calls the __getitem__ function of the Winedataset class on index 0 thats why features = self.x[0] and labels = self.y[0] is how the below code runs.

```python
dataset = WinedataSet()
first_data = dataset[0]                    # __getitem__ is necessarily required to get an indexing on data while using dataloaders and datasets
features,labels = first_data               
print(features,labels)
```

```python
dataloader = DataLoader(dataset,batch_size=4,shuffle=True)          # dataloader internally uses __getitem__ method to load data.
for features,labels in dataloader:
    print(features,labels)
```

if there are n samples; batch size is 4; then u have ceil(n/4) batches. for each epoch every batch is trained hence each epoch has ceil(n/4) training iterations

```python
num_epochs = 3
n_samples = len(dataset)
n_iterations = math.ceil(n_samples/4)

print(n_samples,n_iterations)



for epochs in range(num_epochs):
    for i,(inputs,labels) in (enumerate(dataloader)):
        if ((i+1)%5 == 0):
            print(f'epoch = {epochs+1}/{num_epochs}, step {i+1}/{n_iterations}, inputs {inputs.shape}, labels {labels} ')
```
